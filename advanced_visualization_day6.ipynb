```python
import pandas as pd

data = {
    "Course":[
        "Python","Java","Python",
        "SQL","Java","SQL",
        "Python","Java","SQL","Python"
    ],
    "Gender":[
        "Female","Male","Female",
        "Male","Female","Male",
        "Female","Male","Female","Male"
    ],
    "Marks":[
        85,78,92,67,88,
        73,95,81,89,76
    ],
    "Attendance":[
        90,85,95,75,91,
        80,98,87,92,83
    ]
}

df = pd.DataFrame(data)

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title(
    "Correlation Matrix"
)

plt.show()
```

```python
sns.boxplot(
    x="Course",
    y="Marks",
    data=df
)

plt.show()
```

```python
sns.scatterplot(
    x="Attendance",
    y="Marks",
    hue="Gender",
    data=df
)

plt.show()
```

```python
sns.pairplot(df)

plt.show()
```
