```python
import pandas as pd

data = {
    "Course":[
        "Python","Java","Python",
        "SQL","Java","SQL",
        "Python","Java","SQL","Python"
    ],
    "Marks":[
        85,78,92,67,88,
        73,95,81,89,76
    ]
}

df = pd.DataFrame(data)

df.head()
```

```python
summary = df.groupby(
    "Course"
)["Marks"].mean()

summary.to_excel(
    "automated_course_report.xlsx"
)

print("Report Generated")
```

```python
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt

with PdfPages(
    "student_report.pdf"
) as pdf:

    plt.figure(figsize=(8,5))

    df["Marks"].plot(
        kind="hist"
    )

    pdf.savefig()

    plt.close()

print("PDF Created")
```
